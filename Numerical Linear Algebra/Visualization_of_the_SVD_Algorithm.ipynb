# Visualizating the SVD Algorithm

## Part (a):

```python
# Defining 30 degree angle in radians: 
theta = 30*np.pi/180

# Defining an Empty Vector: 
Y = []

# Generating a 2x12 Matrix:
for i in range(12):
    Y.append([np.cos(i*theta), np.sin(i*theta)])

# Convert to NumPy Array:
Y = np.array(Y)

# Print Y:
print(Y)

# Get Rows and Columns:
rows, cols = Y.T.shape
```

## Part (b):

```python
# Plot the Vectors:
for i, l in enumerate(range(0, cols)):
    xs = [0, Y[i, 0]]
    ys = [0, Y[i, 1]]
    plt.plot(xs, ys)

plt.title("Vectors")
plt.grid()
plt.draw()
plt.show()
```
